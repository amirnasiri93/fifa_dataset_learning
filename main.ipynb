## Feature Engineering

```python
# define new column:
# Age
temp = df[['short_name','age']]
conditions = [
    (temp['age'] <= 20),
    ((temp['age']>20) & (temp['age'] <=25)),
    ((temp['age']>25) & (temp['age'] <=30)),
    (temp['age'] > 30)
]
values = ['< 20', '20 - 25', '26 - 30', '30+' ]

temp['age_tier']=np.select(conditions, values)
temp['age_tier'].value_counts()
```

```python
list(df.columns)
```

```python
# value per overall: 
temp = df[['short_name','value_eur', 'overall']]
temp.head()
temp['value per overall'] = temp['value_eur'] / temp['overall']
temp
```

```python
list(df.columns)
```

```python
# use the attributes of the players to find best attacking 
# players based on some selected attributes:

skills = [
    'skill_dribbling',
    'skill_fk_accuracy',
    'skill_ball_control',
    'dribbling'
]
temp = df[skills]
temp['name'] = df['short_name']
temp['overall'] = df['overall']
temp['skills_score'] = df[skills].mean(axis=1)
temp[['name','skills_score']].sort_values(['skills_score'], ascending = False).head(30)
```

```python
# finding underrated players, calculated based on the skills attributes and the overall rating of players: 
temp['underrated_score'] = temp['skills_score'] - temp['overall']
temp.sort_values('underrated_score', ascending=False)[['name','underrated_score']].head(20)
```

```python
# clubs with highest average overall ratings:
temp = df[['short_name','club_name', 'overall']]
temp.groupby(['club_name'])['overall'].mean().sort_values(ascending=False).head(10).plot(kind='bar', ylabel="overall")
```

```python
temp = df[['short_name','nationality', 'overall']]


df.groupby(['nationality'])['overall'].mean().sort_values(ascending=False).head(10).plot(kind='bar')
```

```python
df[(df['nationality'] == "Brazil")]['overall'].mean()
```

```python
# identify hidden gems; i.e. players with high potential but low overalls:
temp = df[['short_name','overall', 'potential','age']][(df['age']<=22) & (df['overall']<=80) & (df['potential'] >=85)]
temp['hidden_gems_score'] = temp['potential'] - temp['overall']
temp.sort_values(["hidden_gems_score"], ascending=False).head(20)
```

```python
#average player pace per position:
temp = df[['short_name','pace', 'player_positions']]
temp['player_positions'] = temp['player_positions'].apply(lambda a:a.replace(" ","").split(",")[0])
#temp = temp.head(10).sort_values("player_positions")
temp.groupby(["player_positions"])['pace'].mean().fillna(0).sort_values().plot(kind='bar', ylabel='pace')
```

```python
#average player physic per position:
temp = df[['short_name','physic', 'player_positions']]
temp['player_positions'] = temp['player_positions'].apply(lambda a:a.replace(" ","").split(",")[0])
#temp = temp.head(10).sort_values("player_positions")
temp.groupby(["player_positions"])['physic'].mean().fillna(0).sort_values().plot(kind='bar', ylabel='physic')
```

```python
#dribbling
#average player physic per position:
temp = df[['short_name','dribbling', 'player_positions']]
temp['player_positions'] = temp['player_positions'].apply(lambda a:a.replace(" ","").split(",")[0])
#temp = temp.head(10).sort_values("player_positions")
temp.groupby(["player_positions"])['dribbling'].mean().fillna(0).sort_values().plot(kind='bar', ylabel='dribbling')

```
